### **Calculating technical indicators**

We're now going to calculate some important technical indicators: the SMA (Simple Moving Average), the RSI (Relative Strength Index) and the MACD (Moving Average Convergence Divergence).

#### Why Calculate These Technical Indicators?

In trading, stock prices fluctuate constantly, making it difficult to identify trends.  
**Technical indicators** help us analyze the market by providing key insights:  

- **SMA (Simple Moving Average)**  
   - Helps **track the overall trend** by smoothing out price fluctuations.  
   - If the price is **above the SMA**, the trend is **bullish**; otherwise, it's **bearish**.  

- **RSI (Relative Strength Index)**  
   - Measures the **strength of price movement** on a scale from 0 to 100.  
   - An **RSI > 70** indicates **overbought conditions** (potential downturn), while an **RSI < 30** suggests **oversold conditions** (possible rebound).  

- **MACD (Moving Average Convergence Divergence)**  
   - Compares two exponential moving averages (EMA 12 and EMA 26) to detect **trend changes**.  
   - If the MACD crosses **above the signal line**, it’s a **buy signal**; if it crosses below, it’s a **sell signal**.  

These indicators are especially useful for a reinforcement learning model,  
as they allow the agent to analyze market dynamics and optimize trading decisions.


```python
# Calculation of the 20-day SMA (simple moving average)
data['SMA_20'] = data['Close'].rolling(window=20).mean()

# 14-day relative strength index (RSI) calculation
delta = data['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()

rs = gain / loss
data['RSI_14'] = 100 - (100 / (1 + rs))

# MACD and signal line calculation
ema_12 = data['Close'].ewm(span=12, adjust=False).mean()
ema_26 = data['Close'].ewm(span=26, adjust=False).mean()
data['MACD'] = ema_12 - ema_26
data['MACD_signal'] = data['MACD'].ewm(span=9, adjust=False).mean()

# Delete lines containing NaN
print("Nombre de lignes avant dropna():", len(data))
data = data.dropna()
print("Nombre de lignes après dropna():", len(data))


# Display first lines of reorganized data
print("\nThe first lines of reorganized data  :")
print(data.head())
```

## **3. Create trading environment**

### **3.1. Defining the Trading Environment**

The trading environment is a class derived from `gym.Env`. Here are the key steps in its creation:

- **Initialization of environment variables**:

  - **data**: Historical market data.  
  - **balance**: The agent's initial balance.  
  - **position**: The current position in stocks (number of shares held).  
  - **total_assets**: The total value of assets, which is the sum of the balance and the value of the held shares.  

- **Defining the action and observation spaces**:

  - **Action space**: The agent can choose from three actions: **buy (0), sell (1), or hold (2)**.  
  - **Observation space**: The state data, which includes technical indicators: **Close, SMA_20, RSI_14, and MACD**.  

- **reset() method**:  
  This method is called at the beginning of each episode and initializes the environment.  

- **step() method**:  
  This method is called at each step of the episode. It applies the selected action (**buy, sell, or hold**) and updates the state accordingly. The method also returns the reward for the action taken.  

- **render() method**:  
  This method displays the agent’s state, including its **balance, holdings, and profit**.  




```python
import gym
from gym import spaces
import numpy as np
import random
from collections import deque
import matplotlib.pyplot as plt

class StockTradingEnv(gym.Env):
    def __init__(self, data, initial_balance=10000):
        super(StockTradingEnv, self).__init__()
        self.data = data.ffill()  # Fill NaN values using forward-fill method
        self.initial_balance = initial_balance
        self.current_balance = initial_balance
        self.current_step = 0
        self.position = 0  # No initial position
        self.total_assets = self.current_balance

        # Action space: 0 = Buy, 1 = Sell, 2 = Hold
        self.action_space = gym.spaces.Discrete(3)

        # Observation space: Close, SMA, RSI, MACD
        self.observation_space = gym.spaces.Box(low=0, high=np.inf, shape=(4,), dtype=np.float32)

    def _next_observation(self):
      # Extract data from the current step (Close, SMA, RSI, MACD)
        state = self.data.iloc[self.current_step]
        return np.array([
            state['Close'],
            state['SMA_20'],
            state['RSI_14'],
            state['MACD']
        ], dtype=np.float32)

    def reset(self):
        # Reset environment variables at the beginning of each episode
        self.current_balance = self.initial_balance
        self.position = 0
        self.current_step = 0
        self.total_assets = self.current_balance
        print(f"Step 0: Selected action: No movement. Starting observation.")
        return self._next_observation()

    def step(self, action):
        # End of episode condition
        done = self.current_step >= len(self.data) - 1  # Check if max steps reached
        if done:
            return self._next_observation(), 0, done, {}  # End of episode, no reward

        current_price = self.data.iloc[self.current_step]['Close']
        reward = 0

        # Start taking actions only from step 1 (not at step 0)
        if self.current_step == 0:
            print(f"Step {self.current_step}: Selected action: No movement. Starting observation.")
            self.current_step += 1
            return self._next_observation(), 0, done, {}  # No action at step 0

        # Buy action
        if action == 0 and self.current_balance >= current_price:
            amount_to_buy = int(self.current_balance // current_price)  # Number of shares the agent can buy
            if amount_to_buy > 0:
                self.position += amount_to_buy  # Add purchased shares to the position
                self.current_balance -= amount_to_buy * current_price  # Update balance
                print(f"Step {self.current_step}: Selected action: Buy, {amount_to_buy} shares purchased, New balance: {self.current_balance}, New position: {self.position}")
            else:
                print(f"Step {self.current_step}: Selected action: Buy, Insufficient funds to purchase.")

        # Sell action
        elif action == 1 and self.position > 0:
            sale_value = self.position * current_price
            reward = (sale_value - self.initial_balance)  # Calculate profit or loss
            self.current_balance = sale_value  # Update balance with sale proceeds
            self.position = 0  # The agent no longer holds shares after selling
            print(f"Step {self.current_step}: Selected action: Sell, {self.position} shares sold for {sale_value}, New balance: {self.current_balance}, Profit: {reward}")
        elif action == 1 and self.position == 0:
            print(f"Step {self.current_step}: Selected action: Sell, Impossible, no position held.")

        # Hold action
        elif action == 2:
            print(f"Step {self.current_step}: Selected action: Hold, No changes. Current balance: {self.current_balance}")

        # Update the total asset value
        self.total_assets = self.current_balance + self.position * current_price

        # Reward is based on the change in total asset value
        reward = self.total_assets - self.initial_balance

        self.current_step += 1

        # Return the next observation and related info
        return self._next_observation(), reward, done, {}

    def render(self):
        profit = self.total_assets - self.initial_balance
        print(f'Step: {self.current_step}, Balance: {self.current_balance}, Position: {self.position}, Profit: {profit}', flush=True)
```

### **3.2. Create an environment with training data**


```python
# Create an environment with training data
env = StockTradingEnv(train_data)

# Test the environment
state = env.reset()
done = False
total_reward = 0

while not done:
    action = env.action_space.sample()  # Take a random action
    next_state, reward, done, info = env.step(action)
    total_reward += reward
    env.render()  # Display the current state after each action
    state = next_state

display("Total reward:", total_reward)
```

## **4. Testing Different Methods**  
### **4.1. Method 1: Custom DQN with PyTorch**

### **a. Q-Learning Model Definition**

### Defining the Q-Learning Model (Neural Network)
A simple neural network (with three fully connected layers) is used to approximate Q-values in Q-learning. It takes the current state as input and outputs the Q-values for each action (**buy, sell, or hold**).


```python
import torch
import torch.nn as nn
import numpy as np
import gym

class QNetwork(nn.Module):
    def __init__(self, input_size, output_size):
        super(QNetwork, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, output_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

### **b. Agent Training**

### Defining the Training Loop  
During training, the agent selects an action at each step based on the **epsilon-greedy strategy**:  
- With a probability **epsilon (ε)**, it chooses a **random action** (exploration).  
- Otherwise, it selects the action with the **highest Q-value** (exploitation).  


```python
# Environment Initialization
env = StockTradingEnv(train_data)

# Initialize the model, optimizer, and parameters
input_size = 4  # Observation size (Close, SMA, RSI, MACD)
output_size = env.action_space.n  # Number of possible actions
model = QNetwork(input_size, output_size)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Agent parameters
epsilon = 1.0  # Maximum exploration
epsilon_min = 0.01  # Minimum epsilon value
epsilon_decay = 0.995  # Epsilon decay rate
gamma = 0.99  # Discount factor for future rewards
batch_size = 32

# Training the agent
for episode in range(1):  # Using multiple episodes for more complete training
    state = env.reset()
    done = False
    total_reward = 0  # Reset total reward at the start of each episode

    while not done:
        # Define state_tensor before using it in action selection
        state_tensor = torch.tensor(state, dtype=torch.float32).unsqueeze(0)  # Adds a batch dimension (1, 4)

        # Choose action (exploration or exploitation)
        if np.random.rand() < epsilon:
            action = env.action_space.sample()  # Exploration
        else:
            action = torch.argmax(model(state_tensor)).item()  # Exploitation

        # Apply action and retrieve new values
        next_state, reward, done, _ = env.step(action)

        # Add received reward to total reward
        total_reward += reward

        # Convert next_state to Tensor with shape (1, 4)
        next_state_tensor = torch.tensor(next_state, dtype=torch.float32).unsqueeze(0)  # Adds a batch dimension (1, 4)

        # Compute the target (Q-value for the next state)
        with torch.no_grad():
            next_state_q_values = model(next_state_tensor)  # Pass next_state_tensor through the model
            target = reward + gamma * torch.max(next_state_q_values) if not done else reward

        # Compute loss for model weight updates
        predicted = model(state_tensor)[0][action]  # Predicted Q-value for the chosen action
        loss = (target - predicted) ** 2  # Mean Squared Error

        # Update the model
        optimizer.zero_grad()  # Reset gradients
        loss.backward()  # Compute gradients
        optimizer.step()  # Update weights

        state = next_state

        # Call render after each step to display the state during training
        env.render()

    # Call render after each episode to display final episode results
    env.render()  # <---- Displays the final episode state

    # Print the total reward of the episode at the end of each episode
    print(f"Episode {episode+1} - Total Reward: {total_reward}")

    # Reduce epsilon (decreasing exploration)
    epsilon = max(epsilon_min, epsilon * epsilon_decay)
```

### **c. Evaluation on the Test Set**  
Once training is complete, we can test the agent on the test data to evaluate its performance.  


```python
import torch
import matplotlib.pyplot as plt

# Create the test environment with the test data
test_env = StockTradingEnv(test_data)

# Test the DQN agent
portfolio_values_dqn1 = []  # List to store portfolio value at each step
total_test_reward = 0
test_done = False
state = test_env.reset()

# Start evaluation
while not test_done:
    # Convert the current state into a PyTorch tensor
    state_tensor = torch.tensor(state, dtype=torch.float32).unsqueeze(0)

    # Predict the action using the DQN model
    action = torch.argmax(model(state_tensor)).item()

    # Take the action in the environment
    next_state, reward, done, _ = test_env.step(action)

    # Accumulate the reward
    total_test_reward += reward

    # Save the total portfolio value after each step
    portfolio_values_dqn1.append(test_env.total_assets)  # Total assets in your environment

    # Update the state
    state = next_state

    # Display state for debugging (you can disable prints if too verbose)
    print(f"current_step: {test_env.current_step}, done: {done}, total_test_reward: {total_test_reward}")

    # Update the `test_done` flag
    test_done = done

# Test results
print(f"Test Reward DQN: {total_test_reward}")
```

### **4.2 Method 2: Using DQN with the Stable Baselines3 Library**


### **a. Preparing the DQN environment**

```python
from stable_baselines3 import DQN
from stable_baselines3.common.vec_env import DummyVecEnv

# Create a vectorized Gym environment
env = DummyVecEnv([lambda: StockTradingEnv(train_data)])

# Initialize the DQN model
model_dqn = DQN("MlpPolicy", env, verbose=1)

# Train the DQN model (increase timesteps for longer training)
model_dqn.learn(total_timesteps=10000)

# Save the model after training
model_dqn.save("dqn_trading_model")
```

### **b. Test the DQN model and record the portfolio value**

```python
# Load the DQN model
model_dqn = DQN.load("dqn_trading_model")

# Create the custom environment
env = DummyVecEnv([lambda: StockTradingEnv(test_data)])

# Variables to store the results
portfolio_values_dqn2 = []
state = env.reset()  # Start the simulation with the initial state
done = False
total_reward_dqn2 = 0
step_count = 0  # Variable to count the steps

# Training and testing loop
while not done:
    # Predict the action using the DQN model
    action, _states = model_dqn.predict(state)

    # Take the action in the environment and get the results
    state, reward, done, info = env.step(action)

    # Add the reward and portfolio value
    total_reward_dqn2 += reward
    portfolio_values_dqn2.append(env.envs[0].total_assets)

    # Display the results
    print(f"Step: {step_count} - Action: {action} - Reward: {reward} - Total Assets: {env.envs[0].total_assets}")

    # Increase the step count
    step_count += 1

    # If step 254 is reached, stop the episode
    if step_count == 254:
        print(f"Step {step_count} reached, end of episode.")
        done = True  # This will end the loop

# Display the total reward
print(f"Test Reward DQN: {total_reward_dqn2}")
```

```python
# Buy and Hold Strategy - Record portfolio value at each step
portfolio_values_buy_and_hold = []
initial_balance = 10000
buy_and_hold_position = 0

# Perform an initial purchase with the first observation
initial_price = test_data['Close'].iloc[0]
buy_and_hold_position = initial_balance // initial_price

# At each step, calculate the portfolio value
for i in range(1, len(test_data)):
    current_price = test_data['Close'].iloc[i]
    portfolio_value = buy_and_hold_position * current_price + initial_balance - (buy_and_hold_position * initial_price)
    portfolio_values_buy_and_hold.append(portfolio_value)

# Calculate the return of the Buy and Hold strategy
buy_and_hold_reward = portfolio_values_buy_and_hold[-1] - initial_balance
print(f"Buy and Hold Reward: {buy_and_hold_reward}")
```

### **b. Momentum (Buy when the trend is up, sell when it's down)**

```python
# Momentum Strategy - Record portfolio value at each step
portfolio_values_momentum = []
balance = 10000
position = 0

for i in range(1, len(test_data)):
    current_price = train_data['Close'].iloc[i]

    if test_data['MACD'].iloc[i] > 0 and position == 0:
        position = balance // current_price
        balance -= position * current_price
    elif test_data['MACD'].iloc[i] < 0 and position > 0:
        balance += position * current_price
        position = 0

    portfolio_value = balance + position * current_price
    portfolio_values_momentum.append(portfolio_value)

# Calculate the return of the Momentum strategy
momentum_reward = portfolio_values_momentum[-1] - 10000
print(f"Momentum Reward: {momentum_reward}")
```

### **c. Contrarian (Sell when the trend is up, buy when it's down)**

```python
# Contrarian Strategy - Record portfolio value at each step
portfolio_values_contrarian = []
balance = 10000
position = 0

for i in range(1, len(test_data)):
    current_price = test_data['Close'].iloc[i]

    if test_data['MACD'].iloc[i] < 0 and position == 0:
        position = balance // current_price
        balance -= position * current_price
    elif test_data['MACD'].iloc[i] > 0 and position > 0:
        balance += position * current_price
        position = 0

    portfolio_value = balance + position * current_price
    portfolio_values_contrarian.append(portfolio_value)

# Calculate the return of the Contrarian strategy
contrarian_reward = portfolio_values_contrarian[-1] - 10000
print(f"Contrarian Reward: {contrarian_reward}")
```

### **7. Metric calculations**

```python
# Function to calculate volatility
def calculate_volatility(returns):
    if len(returns) > 1:
        return np.std(returns)
    return 0

# Function to calculate maximum drawdown
def calculate_max_drawdown(equity_curve):
    if len(equity_curve) < 2:
        return 0  # Not enough data to calculate drawdown

    peak = equity_curve[0]
    max_drawdown = 0
    for value in equity_curve:
        peak = max(peak, value)
        drawdown = (peak - value) / peak
        max_drawdown = max(max_drawdown, drawdown)
    return max_drawdown

# Function to calculate Sharpe ratio
def calculate_sharpe_ratio(returns, risk_free_rate=0):
    if len(returns) > 1:
        excess_returns = np.array(returns) - risk_free_rate
        return np.mean(excess_returns) / np.std(excess_returns)
    return 0

# Function to calculate Sortino ratio
def calculate_sortino_ratio(returns, risk_free_rate=0):
    if len(returns) > 1:
        downside_returns = [r for r in returns if r < 0]
        downside_volatility = np.std(downside_returns) if downside_returns else 0
        excess_returns = np.array(returns) - risk_free_rate
        return np.mean(excess_returns) / downside_volatility if downside_volatility != 0 else np.nan
    return 0

# Function to calculate Calmar ratio
def calculate_calmar_ratio(returns, equity_curve):
    if len(returns) > 1 and len(equity_curve) > 1:
        annual_return = np.mean(returns)  # Average return
        max_drawdown = calculate_max_drawdown(equity_curve)
        return annual_return / max_drawdown if max_drawdown != 0 else np.nan
    return 0

# For each strategy, portfolios are calculated at each step
# Calculating returns for DQN (Pytorch)
returns_dqn1 = np.diff(portfolio_values_dqn1) / portfolio_values_dqn1[:-1]  # Calculating relative returns
equity_curve_dqn1 = portfolio_values_dqn1  # The equity curve is simply the list of portfolios

# Calculating returns for DQN (Stable Baselines3 Library)
returns_dqn2 = np.diff(portfolio_values_dqn2) / portfolio_values_dqn2[:-1]  # Calculating relative returns
equity_curve_dqn2 = portfolio_values_dqn2  # The equity curve is simply the list of portfolios

# Calculating returns for PPO
returns_ppo = np.diff(portfolio_values_ppo) / portfolio_values_ppo[:-1]
equity_curve_ppo = portfolio_values_ppo

# Calculating returns for Buy and Hold
returns_buy_and_hold = np.diff(portfolio_values_buy_and_hold) / portfolio_values_buy_and_hold[:-1]  # Calculating relative returns
equity_curve_buy_and_hold = portfolio_values_buy_and_hold  # The equity curve is simply the list of portfolios

# Calculating returns for Momentum
returns_momentum = np.diff(portfolio_values_momentum) / portfolio_values_momentum[:-1]  # Calculating relative returns
equity_curve_momentum = portfolio_values_momentum  # The equity curve is simply the list of portfolios

# Calculating returns for Contrarian
returns_contrarian = np.diff(portfolio_values_contrarian) / portfolio_values_contrarian[:-1]  # Calculating relative returns
equity_curve_contrarian = portfolio_values_contrarian  # The equity curve is simply the list of portfolios

# Calculating metrics for each strategy
metrics = {
    'ROI': [],
    'Volatility': [],
    'Max Drawdown': [],
    'Sharpe Ratio': [],
    'Sortino Ratio': [],
    'Calmar Ratio': []
}

# Calculating metrics for DQN (Pytorch)
metrics['ROI'].append((portfolio_values_dqn1[-1] - portfolio_values_dqn1[0]) / portfolio_values_dqn1[0])
metrics['Volatility'].append(calculate_volatility(returns_dqn1))
metrics['Max Drawdown'].append(calculate_max_drawdown(equity_curve_dqn1))
metrics['Sharpe Ratio'].append(calculate_sharpe_ratio(returns_dqn1))
metrics['Sortino Ratio'].append(calculate_sortino_ratio(returns_dqn1))
metrics['Calmar Ratio'].append(calculate_calmar_ratio(returns_dqn1, equity_curve_dqn1))

# Calculating metrics for DQN (Stable Baselines3 Library)
metrics['ROI'].append((portfolio_values_dqn2[-1] - portfolio_values_dqn2[0]) / portfolio_values_dqn2[0])
metrics['Volatility'].append(calculate_volatility(returns_dqn2))
metrics['Max Drawdown'].append(calculate_max_drawdown(equity_curve_dqn2))
metrics['Sharpe Ratio'].append(calculate_sharpe_ratio(returns_dqn2))
metrics['Sortino Ratio'].append(calculate_sortino_ratio(returns_dqn2))
metrics['Calmar Ratio'].append(calculate_calmar_ratio(returns_dqn2, equity_curve_dqn2))

# Calculating metrics for PPO
metrics['ROI'].append((portfolio_values_ppo[-1] - portfolio_values_ppo[0]) / portfolio_values_ppo[0])
metrics['Volatility'].append(calculate_volatility(returns_ppo))
metrics['Max Drawdown'].append(calculate_max_drawdown(equity_curve_ppo))
metrics['Sharpe Ratio'].append(calculate_sharpe_ratio(returns_ppo))
metrics['Sortino Ratio'].append(calculate_sortino_ratio(returns_ppo))
metrics['Calmar Ratio'].append(calculate_calmar_ratio(returns_ppo, equity_curve_ppo))

# Calculating metrics for Buy and Hold
metrics['ROI'].append((portfolio_values_buy_and_hold[-1] - portfolio_values_buy_and_hold[0]) / portfolio_values_buy_and_hold[0])
metrics['Volatility'].append(calculate_volatility(returns_buy_and_hold))
metrics['Max Drawdown'].append(calculate_max_drawdown(equity_curve_buy_and_hold))
metrics['Sharpe Ratio'].append(calculate_sharpe_ratio(returns_buy_and_hold))
metrics['Sortino Ratio'].append(calculate_sortino_ratio(returns_buy_and_hold))
metrics['Calmar Ratio'].append(calculate_calmar_ratio(returns_buy_and_hold, equity_curve_buy_and_hold))

# Calculating metrics for Momentum
metrics['ROI'].append((portfolio_values_momentum[-1] - portfolio_values_momentum[0]) / portfolio_values_momentum[0])
metrics['Volatility'].append(calculate_volatility(returns_momentum))
metrics['Max Drawdown'].append(calculate_max_drawdown(equity_curve_momentum))
metrics['Sharpe Ratio'].append(calculate_sharpe_ratio(returns_momentum))
metrics['Sortino Ratio'].append(calculate_sortino_ratio(returns_momentum))
metrics['Calmar Ratio'].append(calculate_calmar_ratio(returns_momentum, equity_curve_momentum))

# Calculating metrics for Contrarian
metrics['ROI'].append((portfolio_values_contrarian[-1] - portfolio_values_contrarian[0]) / portfolio_values_contrarian[0])
metrics['Volatility'].append(calculate_volatility(returns_contrarian))
metrics['Max Drawdown'].append(calculate_max_drawdown(equity_curve_contrarian))
metrics['Sharpe Ratio'].append(calculate_sharpe_ratio(returns_contrarian))
metrics['Sortino Ratio'].append(calculate_sortino_ratio(returns_contrarian))
metrics['Calmar Ratio'].append(calculate_calmar_ratio(returns_contrarian, equity_curve_contrarian))

# Displaying the results in a DataFrame
metrics_df = pd.DataFrame(metrics, index=['DQN1', 'DQN2', 'PPO', 'Buy and Hold', 'Momentum', 'Contrarian'])
print(metrics_df)
```
